```python
import datetime
current_time = datetime.datetime.now()
year = current_time.strftime("%Y")
month = current_time.strftime("%m")
day = current_time.strftime("%d")

todays_date = f"{year}-{month}-{day}"
todays_date
```

```python
file_name = f"artificialanalysis_environmental_{todays_date}.csv"

url = f"https://raw.githubusercontent.com/Nidhal-Jegham/HowHungryisAIRepo/main/output/{file_name}"
df = pd.read_csv(url)

# Specify the columns to select
selected_cols = [
    "Model",
    "length",
    "Query Length",
    "Hardware",
    "Host",
    "PUE",
    "WUE (Site)",
    "WUE (Source)",
    "Size",
    "Mean Max Energy (Wh)",
    "Mean Min Energy (Wh)",
    "Mean Max Carbon (gCO2e)",
    "Mean Min Carbon (gCO2e)",
    "Mean Max Water (Site, mL)",
    "Mean Min Water (Site, mL)",
    "Mean Max Water (Source, mL)",
    "Mean Min Water (Source, mL)",
    "Energy Consumption of 1 Billion Prompts (MWh)",
    "Carbon Emissions of 1 Billion Prompts (TonsCO2e)",
    "Water Consumption of 1 Billion Prompts (Kiloliter)",
    "Household Energy Equiv. – 1B Prompts (MWh)",
    "People Annual Drinking Water Equiv. – 1B Prompts (kL)",
    "Gasoline Car Equiv. – 1B Prompts (TonsCO2e)",
    "Atlantic Flight Equiv. – 1B Prompts (TonsCO2e)",
]

# Select only the specified columns
df_selected = df[selected_cols]
print(df_selected.head())
df_selected.to_csv(f"../data/{file_name}", index=False)
```
